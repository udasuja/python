    import module_name
과 같은 구문으로 module를 가져온다.

    import math
라 쓰고 math을 입력하고 뒤에 .(마침표)를 찍고 module math가 갖고 있는 기능 중 하나를 적으면 그 기능을 사용할 수 있다.

math.cos()이나 math.sin()과 같이 써 module math의 함수를 사용할 수 있다.

만약 앞에 math를 붙이는 것이 싫고, module내 모든 기능을 가져오는 것이 목적이라면 *기호를 사용한다.

    from math import *
이 구문은 math라는 module 내부의 모든 것을 가져오는 것을 의미한다.

module에는 정말 많은 변수와 함수가 들어 있으지만 활용하고자 하는 기능은 극히 일부일 수 있다. 이때 그 기능들만 가져오는 구문이 다음과 같다.

    from module_name import 가져오고 싶은 변수 또는 함수
import뒤에 쓰는 가져오고 싶은 변수 또는 함수는 여러 개의 변수 또는 함수를 입력할 수도 있다.

module를 가져올 때 이름 충돌이 발생하는 경우 혹은 module의 이름이 너무 길어 사용이 불편할 때가 있을 수 있다. 그때는 다음 같은 as 구문를 사용한다.

    import module_name as 사용하고 싶은 식별자

예
    import math as m
    m.sin(1)

# random

```python
#module random은 랜덤한 값을 생성할 때 사용하는 module이다.
import random
print("module random")

#random(): 0.0<= x < 1.0 사이의 float를 return 한다.
print("- random():",random.random())

#randrange(): 지정한 범위의 int를 return 한다.
# - randrange(max): 0부터 max 사이의 값을 return 한다.
# - randrange(min, max): min부터 max 사이의 값을 return 한다.
print("- randrange(10)",random.randrange(10))

#choice(list): list내부에 있는 element를 랜덤하게 선택한다.
print("- choice([1, 2, 3, 4, 5]):",random.choice([1,2,3,4,5]))

#shuffle(list): list의 element들을 랜덤하게 섞는다.
print("- shuffle([1, 2, 3, 4, 5]):", random.shuffle([1,2,3,4,5]))

#sample(list, k=<숫자>): list의 element중에 k개를 뽑는다.
print("- sample([1, 2, 3, 4, 5], k=2):",random.sample([1,2,3,4,5],k=2))
```

```python
#module datetime은 date, time과 관련된 module로 날짜 형식을 만들 때 자주 사용되는 코드들로 구성되어 있다.
import datetime

#현재 시각을 구하고 출력하기
print("현재 시각 출력")
now=datetime.datetime.now()
print(now.year,"년")
print(now.month,"월")
print(now.day,"일")
print(now.hour,"시")
print(now.minute,"분")
print(now.second,"초")
print()

#시간 출력 방법
print("strftime함수로 시간을 format에 맞춰 출력하기, 다만 한국어등의 문자는 직접적으로 매개변수에 넣을 수 없다.")
output_a=now.strftime("%Y.%m.%d %H:%M:%S")
output_b="{}년 {}월 {}일 {}시 {}분 {}초".format(now.year, \
                                          now.month,\
                                          now.day, \
                                          now.hour, \
                                          now.minute, \
                                          now.second)
output_c=output_a=now.strftime("%Y{} %m{} %d{} %H{} %M{} %S{}").format(*"년월일시분초")  #문자열, list등 앞에 *를 붙이면 element 하나하나가 매개변수로 지정된다.
print(output_a)
print(output_b)
print(output_c)
print()
```

```python
#시간 처리하기
import datetime
now=datetime.datetime.now()

#특정 시간 이후의 시간 구하기
print("현재 시간")
print(now.strftime("%Y{} %m{} %d{} %H{} %M{} %S{}").format(*"년월일시분초"))
print()

print("datetime.timedelta로 시간 더하기") #timedelat()함수를 사용하면 특정한 시간의 이전 또는 이후를 구할 수 있다. 다만 n년 후를 구하는 기능은 없다.
after=now+datetime.timedelta(\
                             weeks=1,\
                             days=1,\
                             hours=1,\
                             minutes=1,\
                             seconds=1)
print(after.strftime("%Y{} %m{} %d{} %H{} %M{} %S{}").format(*"년월일시분초"))
print()

#특정 시간 요소 교체하기, n년 후를 구할 때는 replace()함수를 사용해 아예 날짜 값을 교체하는 것이 일반적이다.
print("now.replace()로 1년 더하기")
output=now.replace(year=(now.year+1))
print(output.strftime("%Y{} %m{} %d{} %H{} %M{} %S{}").format(*"년월일시분초"))
```

```python
from flask import Flask
from urllib import request
from bs4 import BeautifulSoup

#웹 서버를 생성한다.
app=Flask(__name__)
@app.route("/")

def hello():
  #urlopen() 함수로 기상청의 전국 날씨를 읽는다.
  target=request.urlopen("https://www.weather.go.kr/w/weather/forecast/mid-term.do?stnId1=108")

  #BeautifulSoup를 사용해 웹 페이지를 분석한다.
  soup=BeautifulSoup(target,"html.parser")  #BeautifulSoup함수의 매개변수로 HTML 문자열과 "html.parser"라는 문자열을 넣으면 BeautifulSoup이라는 특수한 object를 return한다.

  output=""
  for location in soup.select("location"):
    #내부의 city, wf, tmn, tmx라는 tag를 찾아 출력한다.
    output+="<h3>{}</h3>".format(location.select_on("city").string)
    output+="날씨: {}<br/>".format(location.select_on("wf").string)
    output+="최저/최고 기온: {}/{}".format(location.select_on("tmn").string,location.select_on("tmx").string)
    output+="<hr/>"
  return output
```

module를 살펴보면 library와 framework라는 말을 많이 듣는다. 이 둘은 Inversion of Control(=IoC) 여부에 차이가 있다.

library: 정상적인 control를 하는 module
framework: Inversion of Control이 발생하는 module

그럼 IoC란 무엇이냐?

어떤 module의 내부 기능을 개발자가 직접 호출하는 형태의 module이 library라 할때 module이 개발자가 작성한 코드를 실행하는 형태의 module을 framework라 한다. 즉 개발자가 module의 함수를 호출하는 것이 일반적인 제어 흐름이다. 그런데 이와 반대로 개발자가 만든 함수를 module이 실행하는 것은 제어가 역전된 것이고 이것이 IoC이다.

예를 들어 Flask라는 module을 사용할 때

    from flask import Flask
    app=Flask(__name__)

    @app.route("/")
    def hello():
      return "<h1>Hello World!</h1>"
이렇게만 만들어도 Flask가 내부적으로 웹를 구현해준다. 이러한 module을 framework라 부르고, library는 개발자가 구현하고 싶은 것을 하나하나 전부 다 구현해야 비로서 구현된다.

# decorator라는 기능
@로 시작하는 구문을 python에서는 decorator라 부른다.

decorator는 만드는 방법에 따라 function decorator와 class decorator로 나눌 수 있다.

function decorator: 함수에 사용되는 decorator이다. 즉 대상 함수의 앞뒤에 꾸밀 부가적인 내용을, 혹은 반복할 내용을 decorator로 정의해서 손쉽게 사용할 수 있도록 한 것을 말한다.

```python
#function decorator의 기본
def test(function): #function decorator를 생성한다.
  def wrapper():
    print("인사가 시작되었습니다.")
    function()
    print("인사가 종료되었습니다.")
  return wrapper

#decorator를 붙여 함수를 만든다.
@test
def hello():
  print("hello")

#함수를 호출한다.
hello()
```

위의 코드에 대한 설명

test()함수에서 wrapper()함수를 return하므로, 최종적으로 hello에 함수가 들어가 hello() 형태로 호출할 수 있다.

function decorator를 사용할 때 매개변수 등을 전달할 수 있어 반복되는 구문이 많아질 때 source code의 가독성도 높이고 매우 유용하게 사용할 수 있다. 그리고 functools라는 module를 사용할 수도 있다.

```python
with open("module_a.py","w") as f:
  f.write("variable_a=\"a 모듈의 변수\"")

with open("module_b.py","w") as f:
  f.write("variable_b=\"b 모듈의 변수\"")

with open("main.py","w") as f:
  f.write("""\
import test_package.module_a as a
import test_package.module_b as b

print(a.variable_a)
print(b.variable_b)
""")
```

```python
%run main.py
```

package를 읽을 때 어떤 처리를 수행해야 하거나 package 내부의 module들을 한꺼번에 가져오고 싶을 때가 있다. 이럴 때는 package folder 내부에 \_\_init\_\_.py라는 file을 만들어 사용한다.

그럼 package를 읽어 들일 때 \_\_init\_\_.py를 가장 먼저 실행한다.

따라서 package와 관련된 초기화 처리 등을 할 수 있다. \_\_init\_\_.py에서는 \_\_all\_\_이라는 이름의 list를 만드는데, 이 list에 지정한 module들이 from \<package_name\> import *를 할때 전부 읽어 들여진다.

```python
import os
os.mkdir("test_package")  #folder 생성
os.chdir("/content/test_package/")  #현재 위치 변경

#file 생성
f1=open("module_a.py","w")
f2=open("module_b.py","w")
f3=open("main.py","w")
f1.close()
f2.close()
f3.close()
```

```python
with open("module_a.py","w") as f:
  f.write("variable_a=\"a 모듈의 변수\"")

with open("module_b.py","w") as f:
  f.write("variable_b=\"b 모듈의 변수\"")

with open("__init__.py","w") as f:
  f.write("""\
#main.py에서 from test_package import *로 module를 읽어 들일 때 가져올 module
__all__=["module_a","module_b"]

#package를 읽어 들일 때 처리를 작성할 수도 있다.
print("test_package를 읽어 들였습니다.")
  """)

with open("main.py","w") as f:
  f.write("""\
from test_package import *

print(module_a.variable_a)
print(module_b.variable_b)
""")
```

```python
%run main.py
```
