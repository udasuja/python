program은 data를 처리(가공 혹은 어떠한 유의미한 형태로 변환하는 행위)하는 역할을 한다.
data는 그 data의 기능과 역할에 따라 분류했으며 이렇게 분류된 종류를 data type이라 부른다(string, number, boolean등).
자그마한 data를 data type에 맞게 모으고, 처리 과정을 거쳐 차근 차근 더 큰 data type을 만들어 가면 거대한 program이 만들어진다.
어떤 data의 data type을 알고싶으면 type()함수를 이용한다.
python도 escape character를 지원한다.

```python
print("동해물과 백두산이 마르고 닳도록\n하느님이 보우하사 우리나라 만세\n무궁화 삼천리 화려강산 대한사람\n대한으로 길이 보전하세\n\n\n")

print("-----------------------------------------------")
#python은 여러 줄 문자열이라는 기능을 지원한다.
print("""동해물과 백두산이 마르고 닳도록
하느님이 보우하사 우리나라 만세
무궁화 삼천리 화려강산 대한사람
대한으로 길이 보전하세
""")
print("-----------------------------------------------")
# 다음과 같이 \기호 하나만 사용하면 코드를 쉽게 보려고 줄바꿈 한것이며, 출력할 때는 줄바꿈을 하지 않고 출력되도록 한다.
print("""\
동해물과 백두산이 마르고 닳도록
하느님이 보우하사 우리나라 만세
무궁화 삼천리 화려강산 대한사람
대한으로 길이 보전하세\
""")

print("-----------------------------------------------")
#문자열 연결 연산자: +, 이때 +는 숫자 더하기 연산과는 내부적으로 완전히 다른 수행을 한다. 그렇기에 print("동해물과"+1)은 오류가 난다.
print("동해물과"+"백두산이")

print("-----------------------------------------------")
#문자열 반복 연산자: *, 문자열을 숫자와 *연산자로 연결하여 문자열을 숫자만큼 반복
print("안녕하세요"*7)
print("안녕"+"하세요"*7)
print(("안녕"+"하세요")*7)
print("-----------------------------------------------")
#문자 선택 연산자(indexing): [], 문자열 내부의 문자 하나를 선택하는 연산자.
print("안녕하세요"[0])
print("안녕하세요"[-5])
print("안녕하세요"[4])
print("안녕하세요"[-1])

print("-----------------------------------------------")
#문자열 범위 선택 연산자(slicing): [:], 문자열의 특정 범위를 선택할 때 사용하는 연산자.
hello="안녕하세요"
print("안녕하세요"[1:3])
print("안녕하세요"[1:])
print("안녕하세요"[:3])
print("\n")
print(hello[1:3])
print(hello[1:])
print(hello[:3])
print(hello)
print("-----------------------------------------------")
#문자열의 길이 구할 때는 len()함수 사용
print(len(hello))
```

숫자 연산은 c언어와 대부분 동일한데, 정수 나누기 연산자가 //라는 것과 \**라는 제곱연산자가 존재한다는 차이가 있다. 2^4==2**4

python에서는 변수 선언은 딱히 없고(그냥 변수 이름을 적어주면 그것이 선언이다), 변수를 만들게 되면 모든 data type을 그 변수에 저장할 수 있다. 주의해야 될 점은 변수에 어떤 자료형에 해당하는 값을 넣어주면 그 변수도 변수 내부에 저장된 data의 data type에 대한 변수가 된다.그렇기에 나중에 연산을 할 때 data type을 고려하여 연산해야 한다.

```python
pi=3.14
print(pi,type(pi))
pi="3.14"
print(pi,type(pi))

#명령 프롬프트에서 사용자로부터 data 입력받을 때 사용하는 함수: input(), return 값은 사용자가 입력한 내용이다.return type은 str이다.
da=input("어떤 값을 입력하세요>")  #input함수 괄호 안에 입력한 내용을 prompt string이라 한다.
print(da,type(da))
print("-----------------------------------------------")
#input의 return 값을 int형으로 casting할 수 있다(python에서는 casting은 연산자가 아닌 함수다).
da=int(input("어떤 값을 입력하세요>"))
print(da,type(da))
print("-----------------------------------------------")
```

```python
"""
format()함수는 문자열이 갖고 있는 함수이다. 중괄호{}를 포함한 문자열 뒤에 마침표(.)를 찍고 format()함수를 사용하는데
중괄호의 개수와 format함수 괄호 안 매개변수의 개수는 반드시 같아야한다.

예시:
"{} {}".format(10,20)
이는 앞쪽에 있는 문자열의 {}기호가 format()함수 괄호 안에 있는 매개변수로
차례로 대치되면서 숫자가 문자열이 된다. 즉 10, 20이 "10", "20"이 된다.
"""

string_a="{}".format(10)
print(string_a,type(string_a))

string_b="날짜는 다음과 같습니다. {}년 {}월 {}일".format(2025,9,1)
print(string_b)

string_c="{} {} {}".format(1,"문자열",True)
print(string_c,type(string_c))
#format()함수의 다양한 기능
output_a="{:d}".format(52) #5칸에 출력, 이때 d는 int data type의 정수를 출력하겠다고 직접적으로 지정하는 것이다. 즉 format()함수의 매개변수로 정수가 아닌 값이 오면 오류 발생
print(output_a)

output_b="{:05d}".format(52) #빈칸을 0으로 채우기
output_c="{:05d}".format(-52)
print(output_b)
print(output_c)

output_d="{: d}".format(44)
output_e="{: d}".format(-44)
print(output_d)
print(output_e)

output_f="{:+d}".format(44)
output_g="{:+d}".format(-44) #매개변수가 양수일 때만 출력할 때 접두사에 +를 추가해줌
print(output_f)
print(output_g)

output_h="{:=+5d}".format(-44)  #=는 +기호를 맨 앞으로 밀다는 의미이다.
output_i="{:+05d}".format(-44)
print(output_h)
print(output_i)

#매개변수가 실수일 경우
output_j="{:+015f}".format(52.273)  #15칸 만들고 빈칸들을 0으로 채움
output_k="{:15.3f}".format(52.273)  #15칸 만들고 소수점 아래 3번째 자리까지만 출력, 이때 자동으로 반올림이 진행된다.
print(output_j)
print(output_k)

#매개변수가 실수이지만 소수점에 0만 있는 경우 이 0을 제거
data=52.00
output_l="{:g}".format(data)
print(output_l)

data1=52.05
output_n="{:g}".format(data1)
print(output_n)

#문자열의 알파벳을 소문자에서 대문자로 바꾸는 함수: upper()
string_d="Oh! In case, I don't see yet".upper()
print(string_d)
#문자열의 알파벳을 대문자에서 소문자로 바꾸는 함수: lower()
string_e=string_d.lower()
print(string_e)
#문자열 양옆의 공백을 제거하는 함수: strip()
input_a="""
    안녕하세요
좋은 아침입니다.
"""
print(input_a)
print(input_a.strip())
#문자열 왼쪽의 공백을 제거: lstrip()
#문자열 오른쪽의 공백을 제거: rstrip()

print("-----------------------------------------------")
#문자열이 소문자로만 구성되어 있는지, 알파벳으로만 구성되어 있는지, 숫자로만 구성되어 있는지 등을 확인할 때 is로 시작하는 이름의 함수를 사용한다.
"""
isalnum(): 문자열이 알파벳 혹은 숫자로만 구성되어 있는지 확인
isalpha()
isidentifier()
등
이들 출력은 True혹은 False로 나온다.
"""
print("TrainA10".isalnum())

print("-----------------------------------------------")
#문자열 내부에 특정 문자가 어디에 위치하는지 확인할 때 사용하는 함수
#문자열 왼쪽부터 찾아서 처음 등장하는 위치를 찾는 함수: find()
want_find="안녕안녕하세요".find("안녕")
print("안녕안녕하세요 문자열에서 왼쪽부터 안녕을 찾을 때 index=",want_find)
#문자열 오른쪽부터 찾아서 처음 등장하는 위치를 찾는 함수: rfind()
want_rfind="안녕안녕하세요".rfind("안녕")
print("안녕안녕하세요 문자열에서 오른쪽부터 안녕을 찾을 때 index=",want_rfind)

print("-----------------------------------------------")
#문자열 내부에 어떤 문자열이 있는지 확인하기 위해 in연산자 사용. 출력은 True혹은 False이다.
print("안녕 in 안녕하세요>>","안녕"in"안녕하세요")
print("주무 in 안녕하세요>>","주무"in"안녕하세요")

print("-----------------------------------------------")
#문자열을 특정한 문자로 자를 때 사용하는 함수: split()
a="10 20 30 40 50".split(" ")#이때 괄호 안의 문자열인 공백을 기준으로 자른다.
print("실행 결과는 다음과 같이 list로 출력됨 >>",a)
```
