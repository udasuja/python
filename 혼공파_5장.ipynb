```python
"""
def 함수 이름():
  body

혹은

def 함수 이름(매개변수, 매개변수,...):
  body

이것이 함수의 기본 형태이다.
"""

def print_n_times(value,n):
  for i in range(n):
    print(value)
print_n_times("안녕하세요",5)
```

print()함수와 같이 매개변수를 원하는 만큼 받을 수 있는 함수를 가변 매개변수라 부른다.

    def 함수 이름(매개변수, 매개변수, ...,*가변 매개변수):
      body
이러한 가변 매개변수 뒤에는 일반 매개변수가 올 수 없고, 가변 매개변수는 하나만 사용할 수 있다. 이로써 어디까지가 가변 매개변수인지 구분할 수 있게 된다.

가변 매개변수는 list처럼 사용하면 된다.

```python
def print_n_times(n,*values): #이 함수에서 *values가 가변 매개변수이다.
  for i in range(n):
    for value in values: #values는 list처럼 활용한다.
      print(value)
    print()

print_n_times(3,"안녕하세요","즐거운","파이썬 프로그래밍")
```

기본 매개변수는 '매개변수=값'형태로 되어 있다. 이 매개변수는 특별히 가변 매개변수 뒤에도 올 수 있다.

이러한 기본 매개변수는 함수 호출시 기본 매개변수에 대한 값을 입력하지 않았을 경우 기본 매개변수에 들어가는 기본값을 미리 지정하여 사용하는 매개변수이다.

기본 매개변수 뒤에는 일반 매개변수가 올 수 없다.

```python
def print_n_times(value, n=2):
  for i in range(n):
    print(value)
print_n_times("안녕하세요") #함수 호출 시 n에 대한 값을 입력하지 않아도 자동적으로 2라는 기본값이 매개변수 n에 복사된다.
```

```python
#기본 매개변수가 가변 매개변수보다 앞에 올 때 기본 매개변수의 의미가 사라진다.
def print_n_times(n=2,*values):
  for i in range(n):
    for value in values:
      print(value)
    print()
print_n_times("안녕하세요","즐거운","파이썬 프로그래밍")
#이때는 n에는 "안녕하세요"가 들어가고, values에는 ["즐거운","파이썬 프로그래밍"]이 들어온다.
#그렇기에 range()함수의 매개변수가 정수가 아닌 문자형 data type으로 들어오므로  error가 발생한다
# -> TypeError: 'str' object cannot be interpreted as an integer
```

```python
#가변 매개변수가 기본 매개변수보다 앞에 올 때는 가변 매개변수가 우선된다.
def print_n_times(*values,n=2):
  for i in range(n):
    for value in values:
      print(value)
    print()

print_n_times("안녕하세요","즐거운","파이썬 프로그래밍")

print("-------------------------------------------------")
print_n_times("안녕하세요","즐거운","파이썬 프로그래밍",3)  #3도 가변 매개변수에 들어오는 값으로 취급한다.
print("-------------------------------------------------")

print_n_times("안녕하세요","즐거운","파이썬 프로그래밍",n=3)  #함수 호출 시 기본 매개변수를 따로 취급하고 싶을 때는 기본 매개변수의 이름을 직접적으로 지정해서 값을 입력하면 된다. 이를 키워드 매개변수라 부른다.
```

```python
counter =0

def fibonacci(n):
  global counter  #함수 밖에서 선언한 변수인 counter를 이 함수 안에서도 사용하겠다는 의미이다(외부 변수 참조).
                  #만약 global이라는 키워드가 없으면 함수 내부에서 함수 외부에 있는 변수를 참조하지 못해 error가 발생한다.
  counter +=1
  if n==1:
    return 1
  if n==2:
    return 1
  return fibonacci(n-1)+fibonacci(n-2)

n=int(input("정수 입력> "))
print("fibonacci수열의 {}번째 항의 값: {}".format(n,fibonacci(n)))
print("항의 값을 구할 때까지 더한 횟수: {}".format(counter))
```

위 두 셀을 비교하면 fibonacci라는 함수에 전달되는 매개변수 값이 증가할 수록 그 매개변수에 대한 fibonacci항의 값을 구하기 위한 계산량은 기하급수적으로 늘어난다.

다음은 이 코드를 효율적으로 개선한 코드이다.

```python
#위 두 코드는 재귀를 하면서 같은 값을 구하는 연산을 반복하고 있기에 시간이 오래걸린다.
#그래서 같은 값을 한 번만 계산하도록 코드를 수정한다.
#dictionary를 사용해서 한 번 계산한 값을 저장한다. 이를 memo한다고 표현
#dictionary에 값이 memo되어 있으면 처리를 수행하지 않고 곧바로 memo된 값을 돌려주어 코드의 속도를 빠르게 만든다.
#이러한 기법을 메모화라 한다.
dictionary={
    1:1,
    2:2
}

counter=0
def fibonacci(n):
  global counter
  counter+=1
  if n in dictionary: #memo가 되어 있으면 memo된 값을 return
    return dictionary[n]
  else: #memo가 되어 있지 않으면 값을 구한다.
    output=fibonacci(n-1)+fibonacci(n-2)
    dictionary[n]=output
    return output

n=int(input("정수 입력> "))
print("fibonacci수열의 {}번째 항의 값: {}".format(n,fibonacci(n)))
print("항의 값을 구할 때까지 더한 횟수: {}".format(counter))
```

```python
#함수의 매개변수로 함수 전달
def call_10_times(func):
  for i in range(10):
    func()

def print_hello():
  print("안녕하세요.")

call_10_times(print_hello)  #매개변수로 함수를 전달한다.
```

```python
#함수를 매개변수로 전달하는 대표적인 표준 함수로 map()과 filter()가 있다.
#표준함수는 python이 표준으로 제공하는 함수를 부르는 말이다.
#map(함수, list)  list의 element를 함수에 넣고 return된 값으로 새로운 list를 구성해 주는 함수이다.
#filter(함수, list)   list의 element를 함수에 넣고 return된 값이 True인 것으로, 새로운 list를 구성해주는 함수

counter1=0
def power(item):
  global counter1
  counter1+=1
  return item*item

counter2=0
def under_3(item):
  global counter2
  counter2+=1
  return item<3

list_input_a=[1,2,3,4,5]
output_a=map(power,list_input_a)
print("map()함수의 실행결과")
print("map(power, list_input_a):",output_a)
print("map(power, list_input_a):",list(output_a))
print("map()함수의 호출 횟수:",counter1)
print()
#이는 list의 각 element를 power함수의 매개변수로 각각 받아 power함수를 실행한 뒤
#그렇게 실행하여 얻은 power함수의 모든 return값을 차례대로 map함수가 받아 새로운 list로 구성해준다.

output_b=filter(under_3,list_input_a)
print("filter()함수의 실행결과")
print("filter(under_3, output_b):",output_b)
print("filter(under_3, output_b):",list(output_b))
print("filter()함수의 호출 횟수:",counter2)
#이는 list의 각 element를 under_3함수의 매개변수로 각각 받아 under_3함수를 실행한 뒤
#filter함수는 under_3함수가 출력한 모든 return 값 중 True를 출력한 경우를 모아 그 경우의 under_3함수의 매개변수만으로 새로운 list를 구성해준다.
```

위 코드를 보면 map()함수의 실행결과로는 <map object at 0x7987b2856f20>가

filter()함수의 실행결과로는 <filter object at 0x7987b2854ac0>가 나온다.

이를 generator라 부른다.


위처럼 매개변수로 함수를 전달하기 위해 함수 구문을 작성하는 것도 번거롭고, 코드 공간 낭비라는 생각이 든다. 그래서 lambda라는 개념이 나온 것이다.

lambda는 간단한 함수를 쉽게 선언하는 방법으로

    lambda 매개변수: 리턴값
형태로 만든다.

lambda를 사용하면 def 키워드로 선언했던 함수를 lambda로 바꾸고, return 키워드를 따로 쓰지 않는다라는 차이가 있다.

다음과 같이 매개변수가 여러 개인 lambda도 만들 수 있다.

    lambda x, y: x * y

```python
power=lambda x:x*x  #power()함수 선언
under_3=lambda x:x<3  #under_3()함수 선언

list_input_a=[1,2,3,4,5]

output_a=map(power,list_input_a)
print("map()함수의 실행결과")
print("map(power,list_input_a):",output_a)
print("map(power,list_input_a):",list(output_a))
print()

output_b=filter(under_3,list_input_a)
print("filter()함수의 실행결과")
print("filter(under_3,output_b):",output_b)
print("filter(under_3, output_b):",list(output_b))
```

```python
#이러한 lambda는 함수의 매개변수에 곧바로 넣을 수도 잇다.

list_input_a=[1,2,3,4,5]

output_a=map(lambda x:x*x,list_input_a) #power함수를 선언하지도 않고, 매개변수로 바로 넣음
print("map()함수의 실행결과")
print("map(power,list_input_a):",output_a)
print("map(power,list_input_a):",list(output_a))
print()

output_b=filter(lambda x:x<3,list_input_a)
print("filter()함수의 실행결과")
print("filter(under_3, output_b):",output_b)
print("filter(under_3, output_b):",list(output_b))
```

# 파일 처리

파일은 크게 text file과 binary file로 나뉜다.
file를 열 때는 open()함수를 사용한다.

    file object=open(문자열: 파일경로, 문자열: 읽기 모드)
이때 모드는 w(=새로 write), a(=뒤에 이어서 write), r(=read mode)이 있다.

file를 닫을 때는 close()함수를 사용한다.

    file object.close()


```python
file=open("basic.txt","w")  #basic.txt파일이 없으면 새로 만든다. 어차피 w라는 mode로 접근하기 때문
file.write("Hello Python Programming...!")
file.close()
```

program이 길어지면 open()함수와 close()함수 사이에 많은 코드가 들어간다.

코드가 길어지면서 자연스럽게 file을 open하고는 close하지 않는 실수를 범할 수 있다. 그렇기에 이런 실수를 방지하기 위해 with라는 keyword 기능이 존재한다.

    with open(문자열: 파일 경로, 문자열: 모드) as file object:
        body
with라는 keyword에 들여쓰기 된 body라는 코드가 전부 동작되고 with를 빠져나올 때 열린 file은 자동으로 close된다.

이러한 with 키워드는 결국 stream를 열고 닫는 실수를 줄이고자 만들어진 구문으로 network로 흐르는 길을 열고 닫을 때에도 사용된다.

```python
with open("basic.txt","w") as file:
  file.write("Hello Python Programming...!")
```

file을 읽을 때는 read()함수를 사용한다.

    file object.read()
file을 열고 file object의 read()함수를 호출하기만 하면 내부에 있는 data를 모두 읽을 수 있다.

```python
with open("basic.txt","r") as file:
  contents=file.read()
print(contents)
```

CSV file은 한 행이 하나의 data를 나타내며, 각각의 행은 쉼표를 사용해 data를 구분한다. 첫 번째 행에 header를 넣어 각 data가 무엇을 나타내는지 설명해 줄 수 있다.

```python
#랜덤하게 1000명의 키와 몸무게 만들기
import random

hanguls=list("가나다라마바사아자차카타파하")
with open("info.txt","w") as file:
  for i in range(1000):
    #랜덤한 값으로 변수를 생성한다.
    name=random.choice(hanguls)+random.choice(hanguls)
    weight=random.randrange(40,100)
    height=random.randrange(140,200)
    file.write("{}, {}, {}\n".format(name, weight, height))

#반복문으로 file 한 줄씩 읽기
with open("info.txt","r") as file:
  for line in file:
    (name, weight, height)=line.strip().split(", ") #변수를 선언한다.
    if(not name) or (not weight) or (not height): #data가 문제없는지 확인한다.
      continue
    bmi=int(weight)/(int(height)*int(height))
    result=""
    if 25<=bmi:
      result="과체중"
    elif 18.5<=bmi:
      result="정상 체중"
    else:
      result="저체중"

    print('\n'.join([
        "이름: {}",
        "몸무게: {}",
        "키: {}",
        "BMI: {}",
        "결과: {}"
    ]).format(name, weight, height, bmi, result))
    print()
```

# generate
generator는 iterator를 직접 만들 대 사용하는 코드이다. 함수 내부에 yield라는 keyword를 사용하면 해당 함수는 generator 함수가 되며, 일반 함수와는 달리 함수를 호출해도 함수 내부의 코드가 실행되지 않는다.

이러한 generator 함수는 generator를 리턴한다.

```python
def test():
  print("함수가 호출되었습니다.")
  yield "test"

print("A지점 통과")
test()

print("B지점 통과")
test()
print(test())
```

출력된 <generator object test at 0x7987b236f280>는 generator object이다. 이 object는 next()함수를 사용해 함수 내부의 코드를 실행한다. 이때 yield라는 keyword 부분까지만 실행하며, next()함수의 리턴값으로 yield라는 keyword 뒤에 입력한 값이 출력된다.

next()함수를 호출한 이후 yield라는 keyword를 만나지 못하고 함수가 끝나면 StopIteration라는 예외가 발생한다.

이처럼 generator object는 메모리의 효율성을 위해 함수의 코드를 조금씩 실행할 때 사용한다.

```python
def test():
  print("A 지점 통과")
  yield 1
  print("B 지점 통과")
  yield 2
  print("C 지점 통과")

output=test()

print("D지점 통과")
a=next(output)
print(a)

print("E 지점 통과")
b=next(output)
print(b)

next(output)
```
